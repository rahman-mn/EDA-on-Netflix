```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

sns.set(style="whitegrid", palette="Set2")
```

```python
df = pd.read_csv(r"C:\Users\user\Downloads\archive\NetFlix.csv")


print("Rows, Columns", df.shape)
df.head()
```

```python
df.info()

df.describe(include='all').T
```

```python
# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Clean text fields
text_cols = ['title', 'director', 'cast', 'country', 'rating', 'genres', 'description']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().replace({'nan': np.nan})

# Split duration into numeric and unit parts
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract(r'([A-Za-z]+)')
```

```python
# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Clean text fields
text_cols = ['title', 'director', 'cast', 'country', 'rating', 'genres', 'description']
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().replace({'nan': np.nan})

# Split duration into numeric and unit parts
# First convert duration to string and handle NaN values
df['duration'] = df['duration'].astype(str)
df.loc[df['duration'] == 'nan', 'duration'] = np.nan

# Now extract numeric and unit parts only for non-null values
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract(r'([A-Za-z]+)')
```

```python
df.isna().sum().sort_values(ascending=False)

```

```python
msno.bar(df)
plt.show()
```

```python
sns.countplot(data=df, x='type')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
plt.title("Number of Titles Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
df['rating'].value_counts().plot(kind='barh', figsize=(8,5))
plt.title("Distribution of Ratings")
plt.show()
```

```python
sns.histplot(df[df['type']=="Movie"]['duration_num'], bins=30, kde=True)
plt.title("Movie Durations (minutes)")
plt.show()

sns.countplot(data=df[df['type']=="TV Show"], x='duration_num')
plt.title("TV Shows by Number of Seasons")
plt.show()
```

```python
genres = df['genres'].str.split(',').explode().str.strip().value_counts().head(15)
genres.plot(kind='barh', figsize=(8,6))
plt.title("Top 15 Genres")
plt.show()
```

```python
countries = df['country'].dropna().str.split(',').explode().str.strip().value_counts().head(15)
countries.plot(kind='barh', figsize=(8,6))
plt.title("Top 15 Content-Producing Countries")
plt.show()
```

```python
sns.countplot(data=df, x='rating', hue='type')
plt.xticks(rotation=45)
plt.title("Rating Distribution by Type")
plt.show()
```

```python
sns.histplot(data=df, x='release_year', hue='type', multiple='stack', bins=40)
plt.title("Release Year vs Type")
plt.show()
```

```python
top10 = df['country'].dropna().str.split(',').explode().str.strip().value_counts().head(10).index
filtered = df[df['country'].fillna('').apply(lambda x: any(c in x for c in top10))]
sns.countplot(data=filtered, y='country', hue='type', order=top10)
plt.title("Top Countries by Content Type")
plt.show()
```

```python
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
plt.title("Titles Added to Netflix Over the Years")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english', max_features=25)
X = cv.fit_transform(df['description'].dropna())

pd.Series(X.toarray().sum(axis=0), index=cv.get_feature_names_out()) \
    .sort_values().plot(kind='barh', figsize=(8,6))

plt.title("Most Common Words in Descriptions")
plt.show()
```

```python
df['num_cast'] = df['cast'].apply(lambda x: len(str(x).split(',')) if pd.notna(x) else 0)
df['num_genres'] = df['genres'].apply(lambda x: len(str(x).split(',')) if pd.notna(x) else 0)
df['is_recent_release'] = (df['release_year'] >= 2018).astype(int)

df[['title', 'type', 'release_year', 'num_cast', 'num_genres', 'is_recent_release']].head()
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
print("✅ Cleaned data saved successfully.")
```
